```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
path = 'Data/Medals.xlsx'
medals = pd.read_excel(path, engine="openpyxl")
medals.head(10)
```

```python
total_medals = medals['Gold'] + medals['Silver'] + medals['Bronze']
medals['Total'] = total_medals
medals.head()
```

```python
plt.figure(figsize=(10,7))
pie=plt.pie(x=medals["Total"],autopct='%1.02f%%',radius=4)
plt.legend(pie[0],labels=medals["Team/NOC"], bbox_to_anchor=(1,0), loc="best", bbox_transform=plt.gcf().transFigure)
plt.title("Total medals won by the participating Countries represented as a percentage",fontsize=20,fontweight="bold",loc="left")
```
