```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'Data/Teams.xlsx'
df = pd.read_excel(path, engine="openpyxl")
df.head(5)
```

```python
df.tail(5)
```

```python
plt.figure(figsize=(20,3))
sns.barplot(x=df.Discipline.value_counts().index,
            y=df.Discipline.value_counts().values)
plt.xlabel('Discipline')
plt.ylabel('Frequency')
plt.title('Bar Plot of Discipline')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,3))
sns.barplot(x=df.NOC.value_counts().index,
            y=df.NOC.value_counts().values)
plt.xlabel('NOC')
plt.ylabel('Frequency')
plt.title('Barplot of NOC')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,3))
sns.barplot(x=df.Event.value_counts().index,
            y=df.Event.value_counts().values)
plt.xlabel('Event')
plt.ylabel('Frequency')
plt.title('Bar Plot of Event')
plt.xticks(rotation=90)
plt.show()
```
