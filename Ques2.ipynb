```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
athletes = pd.read_excel('Data/Athletes.xlsx')
coaches = pd.read_excel('Data/Coaches.xlsx')
gender = pd.read_excel('Data/EntriesGender.xlsx')
medals = pd.read_excel('Data/Medals.xlsx')
teams = pd.read_excel('Data/Teams.xlsx')
```

```python
def missing_val_analysis(df):
    sum=df.isnull().sum()
    percentage=sum/len(df)
    msno.bar(df)
    plt.show()
    return sum, percentage
```

```python
missing_val_analysis(athletes)
```

```python
missing_val_analysis(coaches)
```

```python
missing_val_analysis(gender)
```

```python
missing_val_analysis(medals)
```

```python
missing_val_analysis(teams)
```
